```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("istherecorrelation.csv", sep= ";")
plt.plot(df["WO [x1000]"], df["NL Beer consumption [x1000 hectoliter]"])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Amount of WO students in NL [x1000]")
plt.ylabel("NL Beer consumption [x1000 hectoliter]")
plt.title("Amount of beer consumed in NL compaired to amount of WO students in NL")
plt.show()
```
